```python
from pyproj import Proj, transform

# Función para obtener el EPSG de UTM basado en latitud y longitud
def obtener_epsg_utm(lat, lon):
    proj_utm = Proj(proj='utm', zone=int((lon + 180) / 6) + 1, ellps='WGS84', south=(lat < 0))
    epsg_utm = proj_utm.crs.to_epsg()
    return epsg_utm

# Ejemplo de uso
latitud = 21.161785  # Ejemplo con la latitud de Nueva York
longitud = -86.851047  # Ejemplo con la longitud de Nueva York
epsg_utm_code = obtener_epsg_utm(latitud, longitud)
print(f"El código EPSG de UTM para la latitud {latitud} y longitud {longitud} es: {epsg_utm_code}")
```

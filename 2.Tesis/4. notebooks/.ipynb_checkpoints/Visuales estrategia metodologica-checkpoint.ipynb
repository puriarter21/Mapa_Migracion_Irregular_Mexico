```python
dfvalores1 = dfvalores[['nommun', 'estado', 'region', 'id_poligon', 'eventos', 'geometry', 'distancia_km', 'inm_30km_adjusted', 'latitud']]
```

```python
import geopandas as gpd
from shapely.geometry import Point

# Paso 1: Verificar si 'geometry' ya contiene objetos Point. Si no, convertir.
dfvalores1['geometry'] = dfvalores1['geometry'].apply(lambda x: x if isinstance(x, Point) else Point(map(float, x.replace('POINT (', '').replace(')', '').split())))
dfvalores1 = gpd.GeoDataFrame(dfvalores1, geometry='geometry', crs="EPSG:4326")

# Paso 2: Crear un polígono de 30 km alrededor del punto (asegurarse de la proyección correcta)
dfvalores1['geometry_buffer_30km'] = dfvalores1['geometry'].to_crs(epsg=3857).buffer(30000).to_crs(epsg=4326)

# Paso 3: Verificar si inm_opt_conc1['geometry'] también ya contiene objetos Point
inm_opt_conc1['geometry'] = inm_opt_conc1['geometry'].apply(lambda x: x if isinstance(x, Point) else Point(map(float, x.replace('POINT (', '').replace(')', '').split())))
inm_opt_conc1 = gpd.GeoDataFrame(inm_opt_conc1, geometry='geometry', crs="EPSG:4326")

# Paso 4: Contar cuántos puntos de inm_opt_conc1 están dentro del polígono de 30 km de cada punto en dfvalores1
def contar_puntos_dentro_poligono(fila):
    poligono = fila['geometry_buffer_30km']
    conteo = inm_opt_conc1[inm_opt_conc1['geometry'].within(poligono)].shape[0]
    return conteo + 1  # Sumar 1 al resultado

# Crear la nueva columna 'inm_30_op' con el conteo
dfvalores1['inm_30_op'] = dfvalores1.apply(contar_puntos_dentro_poligono, axis=1)




```

```python
dfvalores1.head()
```

```python
import numpy as np

# Paso 1: Calcular el logaritmo natural para eventos / inm_30km_adjusted
dfvalores1['log_eventos_per_inm_30km'] = np.log1p(dfvalores1['eventos'] / dfvalores1['inm_30km_adjusted'])

# Paso 2: Calcular el logaritmo natural para eventos / inm_30_op
dfvalores1['log_eventos_inm_op'] = np.log1p(dfvalores1['eventos'] / dfvalores1['inm_30_op'])

# Verificar el resultado para las nuevas columnas calculadas
print(dfvalores1[['nommun', 'eventos', 'inm_30km_adjusted', 'inm_30_op', 'log_eventos_per_inm_30km', 'log_eventos_inm_op']].head())
```

```python
# Paso 1: Encontrar el valor máximo entre todas las filas para ambas columnas de logaritmos
max_log_value = max(dfvalores1['log_eventos_per_inm_30km'].max(), dfvalores1['log_eventos_inm_op'].max())

# Paso 2: Normalizar ambas columnas dividiendo por el valor máximo
dfvalores1['norm_log_eventos_per_inm_30km'] = dfvalores1['log_eventos_per_inm_30km'] / max_log_value
dfvalores1['norm_log_eventos_inm_op'] = dfvalores1['log_eventos_inm_op'] / max_log_value

# Verificar los resultados para asegurarse de que la normalización se realizó correctamente
print(dfvalores1[['nommun', 'log_eventos_per_inm_30km', 'log_eventos_inm_op', 'norm_log_eventos_per_inm_30km', 'norm_log_eventos_inm_op']].head())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Reemplazar valores infinitos por NaN
dfvalores1.replace([np.inf, -np.inf], np.nan, inplace=True)

# Eliminar filas con valores NaN en las columnas relevantes
dfvalores_cleaned = dfvalores1.dropna(subset=['distancia_km', 'norm_log_eventos_per_inm_30km', 'norm_log_eventos_inm_op', 'region'])

# Ordenar el DataFrame por 'distancia_km' de menor a mayor
dfvalores_sorted = dfvalores_cleaned.sort_values(by='distancia_km')

# Configuración de Seaborn
sns.set(style="whitegrid")

# Crear el grid de subplots, dividido por 'region'
g = sns.FacetGrid(dfvalores_sorted, col='region', col_wrap=2, sharex=True, sharey=True, height=5, aspect=2)

# Función para graficar las dos líneas
def plot_lines(data, **kwargs):
    # Línea sólida más delgada para 'norm_log_eventos_per_inm_30km'
    sns.lineplot(x='distancia_km', y='norm_log_eventos_per_inm_30km', data=data, marker='o', label='Eventos por Inm 30km', color='b', lw=1)
    
    # Línea punteada más delgada para 'norm_log_eventos_inm_op'
    sns.lineplot(x='distancia_km', y='norm_log_eventos_inm_op', data=data, marker='o', label='Eventos por Inm OP', linestyle='--', color='g', lw=1)
    
    # Agregar promedios y sombras
    mean_30km = data['norm_log_eventos_per_inm_30km'].mean()
    mean_inm_op = data['norm_log_eventos_inm_op'].mean()
    std_30km = data['norm_log_eventos_per_inm_30km'].std()
    std_inm_op = data['norm_log_eventos_inm_op'].std()

    # Líneas de promedio
    plt.axhline(mean_30km, color='b', linestyle='--', lw=1.5, label='Promedio Eventos Inm 30km')
    plt.axhline(mean_inm_op, color='g', linestyle='--', lw=1.5, label='Promedio Eventos Inm OP')

    # Sombra para desviación estándar
    plt.fill_between(data['distancia_km'], mean_30km - std_30km, mean_30km + std_30km, color='b', alpha=0.1)
    plt.fill_between(data['distancia_km'], mean_inm_op - std_inm_op, color='g', alpha=0.1)

    # Calcular la diferencia entre las dos series
    data['diferencia'] = abs(data['norm_log_eventos_per_inm_30km'] - data['norm_log_eventos_inm_op'])
    
    # Identificar el punto con la mayor variación (diferencia)
    max_variacion_idx = data['diferencia'].idxmax()
    max_row_variacion = data.loc[max_variacion_idx]
    
    # Añadir anotación en el punto con la mayor variación
    plt.annotate('Mayor Variación', xy=(max_row_variacion['distancia_km'], max_row_variacion['norm_log_eventos_per_inm_30km']),
                 xytext=(5, 5), textcoords='offset points', arrowprops=dict(arrowstyle='->', color='red'), color='red')

# Mapear la función a cada subplot
g.map_dataframe(plot_lines)

# Añadir títulos y etiquetas
g.set_axis_labels('Distancia (km)', 'Log normalizado (0-1)')
g.set_titles(col_template='{col_name} Region')

# Ajustar el layout
plt.subplots_adjust(top=0.85)
g.fig.suptitle('Eventos Normalizados por Región vs Distancia', fontsize=20)

# Añadir leyenda debajo de los gráficos
handles, labels = g.axes[0].get_legend_handles_labels()
g.fig.legend(handles, labels, loc='lower center', ncol=2, title='Variables', bbox_to_anchor=(0.5, -0.1))

# Mostrar la gráfica
plt.show()
```

```python
estado_region_map = osc.dropna(subset=['region']).groupby('estado')['region'].first().to_dict()

# Función para completar las regiones faltantes
def llenar_region(row):
    if pd.isna(row['region']):  # Si el valor en region es NaN
        estado = row['estado']
        if estado in estado_region_map:
            row['region'] = estado_region_map[estado]
    return row

# Aplicar la función a todo el DataFrame
osc_corregido = osc.apply(llenar_region, axis=1)

# Mostrar el resultado para verificar
print(osc_corregido[['estado', 'region']].head())

# Guardar el DataFrame corregido si es necesario
# osc_corregido.to_csv('archivo_corregido.csv', index=False)
```

```python
# Obtener los valores únicos de las columnas 'municipio' y 'estado'
unique_municipios = osc['municipio'].unique()
unique_estados = osc['estado'].unique()

# Mostrar los valores únicos de ambas columnas
print("Valores únicos en la columna 'municipio':")
print(unique_municipios)

print("\nValores únicos en la columna 'estado':")
print(unique_estados)

# Opción adicional: Mostrar valores que no coinciden entre municipio y estado
print("\nValores en 'municipio' que parecen un 'estado':")
print([m for m in unique_municipios if m in unique_estados])

print("\nValores en 'estado' que parecen un 'municipio':")
print([e for e in unique_estados if e in unique_municipios])
```

```python
# Definir listas correctas de estados y municipios
estados_validos = ['Guerrero', 'Veracruz', 'Coahuila', 'Sonora', 'Aguascalientes', 'Sinaloa', 
                   'Tamaulipas', 'D.F.', 'Michoacan', 'Tlaxcala', 'Nuevo Leon', 'Chiapas', 
                   'Hidalgo', 'Mexico', 'Jalisco', 'Quintana Roo', 'Campeche', 'Tabasco', 
                   'Guanajuato', 'Chihuahua', 'Oaxaca', 'San Luis Potosi', 'Colima', 
                   'Queretaro', 'Baja California Sur', 'Morelos', 'Durango', 
                   'Baja California', 'Puebla', 'Nayarit', 'Yucatan', 'Zacatecas']

# Revisar cada fila y corregir si los valores están intercambiados
def corregir_intercambio(row):
    municipio = row['municipio']
    estado = row['estado']
    
    # Si el municipio está en la lista de estados y el estado en la lista de municipios, los intercambiamos
    if municipio in estados_validos and estado not in estados_validos:
        row['municipio'], row['estado'] = estado, municipio
    return row

# Aplicar la corrección a todo el DataFrame
osc_corregido = osc.apply(corregir_intercambio, axis=1)

# Guardar o ver las primeras filas para verificar la corrección
print(osc_corregido[['municipio', 'estado']].head())
```

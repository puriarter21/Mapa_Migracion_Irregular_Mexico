```python
# Supongamos que tienes dos dataframes: df1 y df2
inm_opt_conc= pd.concat([inm, inm_opt], ignore_index=True)
```

```python
# Iterar por todas las filas con etiquetas == 'original'
for idx_org, fila_org in inm_opt_conc[inm_opt_conc['etiquetas'] == 'original'].iterrows():
    # Lista para almacenar las filas 'optimas' dentro del rango de 30 km
    filas_cercanas = []

    # Iterar por todas las filas con etiquetas == 'optimas'
    for idx_opt, fila_opt in inm_opt_conc[inm_opt_conc['etiquetas'] == 'optimas'].iterrows():
        distancia = calcular_distancia(fila_org.geometry, fila_opt.geometry)
        
        # Verificar si la distancia es menor o igual a 30 km
        if distancia <= 60:
            filas_cercanas.append((idx_opt, distancia))
    
    # Si hay filas cercanas, cambiar la etiqueta de la fila original a 'org/elim'
    if len(filas_cercanas) > 0:
        inm_opt_conc.at[idx_org, 'etiquetas'] = 'org/elim'

        # Encontrar la fila mÃ¡s cercana dentro del rango
        fila_mas_cercana = min(filas_cercanas, key=lambda x: x[1])
        
        # Cambiar la etiqueta de la fila mÃ¡s cercana a 'org/opt'
        inm_opt_conc.at[fila_mas_cercana[0], 'etiquetas'] = 'org/opt'
```

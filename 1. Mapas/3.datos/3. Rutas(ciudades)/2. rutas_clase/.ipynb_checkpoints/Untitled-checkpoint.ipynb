```python
# Obtener los valores únicos de la columna 'codigo_ruta'
codigos_unicos = datos_csv['codigo_ruta'].unique()

# Crear un diccionario para almacenar los sub-DataFrames
sub_dfs = {}

for codigo in codigos_unicos:
    # Crear un sub-DataFrame para cada código y almacenarlo en el diccionario
    sub_dfs[codigo] = datos_csv[datos_csv['codigo_ruta'] == codigo]


```

```python
sub_dfs['S'].head()
```

### Guardamos cada archivo

```python
# Ruta base para guardar los archivos
ruta_base = '/Users/pablouriarte/Documents/1. Expediente Tec de Monterrey/1.Tesis/Mapa_Migracion_Irregular_Mexico/3. mapas/3. Rutas/2. rutas_clase/'

# Iteramos sobre el diccionario de sub-DataFrames
for codigo, sub_df in sub_dfs.items():
    # Construimos el nombre del archivo usando el código de ruta
    nombre_archivo = f"{codigo}_rutas_clase.csv"
    
    # Construimos la ruta completa del archivo
    ruta_completa = ruta_base + nombre_archivo
    
    # Guardamos el sub-DataFrame en un archivo CSV
    sub_df.to_csv(ruta_completa, index=False)
    
    # Mensaje de confirmación
    print(f"Archivo {nombre_archivo} guardado con éxito en {ruta_completa}")
```

```python
import pandas as pd  # Library for data manipulation and analysis
import matplotlib.pyplot as plt  # Library for creating visualizations
import seaborn as sns  # Library for statistical data visualization
```

```python
bd_tw = pd.read_csv('/Users/pablouriarte/Documents/1. Expediente Tec de Monterrey/1.Tesis/otros/KML Servicios/lugares.csv')
```

```python
#minusculas columnas
bd_tw.columns = bd_tw.columns.str.lower()

#bd_tw.head()
```

```python
bd_tw.head()
```

```python
#
```

```python
#Quitar todas las columnas que no tienen info
# Seleccionar solo las columnas 'description', 'geometry_str', 'tipo' y 'name'
bd_tw = bd_tw[['description', 'geometry_str', 'tipo', 'name']]

#Igual lo podemos hacer con las cantidad de nan dada las caracteristicas de la base de datos
# Calcula el umbral
threshold = len(bd_tw) - 100

# Elimina las columnas que tienen menos valores no nulos que el umbral
bd_tw = bd_tw.dropna(axis=1, thresh=threshold)
```

```python
#Para extraer las direcciones
# Separa la columna 'description' en varias columnas
descriptions = bd_tw['description'].str.split('<br>', expand=True)

# Cambia el nombre de las columnas a 'descripcion_uno', 'descripcion_dos', etc.
descriptions.columns = [f'descripcion_{i+1}' for i in range(descriptions.shape[1])]

# Une las nuevas columnas al DataFrame original
bd_tw = pd.concat([bd_tw, descriptions], axis=1)
```

Esto no hace la separacion correctamente, hay que ver alguna manera de extraer las direcciones.
Sin embargo, por el momento solo necesito las coordenadas.

```python
#bd_tw.head(50)
```

```python
bd_tw_coord = bd_tw[["name", "tipo", "geometry_str"]]
```

```python
bd_tw_coord.head()
```

```python
# Extraer las coordenadas de la columna 'geometry_str'
bd_tw_coord[['latitud', 'longitud']] = bd_tw_coord['geometry_str'].str.extract('c\((.*), (.*), .*')

# Convertir las columnas 'latitud' y 'longitud' a float
bd_tw_coord[['latitud', 'longitud']] = bd_tw_coord[['latitud', 'longitud']].astype(float)

# Eliminar la columna 'geometry_str'
bd_tw_coord = bd_tw_coord.drop('geometry_str', axis=1)

```

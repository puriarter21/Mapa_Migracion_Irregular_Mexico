```python
# Libraries
import pandas as pd  # Library for data manipulation and analysis
import matplotlib.pyplot as plt  # Library for creating visualizations
import seaborn as sns  # Library for statistical data visualization

import folium
```

```python
bd_tw = pd.read_csv('/Users/pablouriarte/Documents/1. Expediente Tec de Monterrey/1.Tesis/3. Datos/pruebainstituciones.csv')
```

```python
bd_tw.head()
```

```python
# Dividir la columna "Coordenadas" en dos columnas: 'Latitud' y 'Longitud'
bd_tw[['Latitud', 'Longitud']] = bd_tw['Coordenadas'].str.split(',', expand=True)

# Eliminar la columna 'Coordenadas' y 'Unnamed: 3'
bd_tw.drop(columns=['Coordenadas', 'Unnamed: 3'], inplace=True)

# Imprimir el DataFrame limpio
print(bd_tw)
```

```python
bd_tw.head()
```

```python
import pandas as pd

# Suponiendo que ya tienes el DataFrame "bd_tw"

# Reemplazar el símbolo de grados y las letras en las columnas "Latitud" y "Longitud"
bd_tw['Latitud'] = bd_tw['Latitud'].str.replace('° N', '').str.replace('° S', '')
bd_tw['Longitud'] = bd_tw['Longitud'].str.replace('° E', '').str.replace('° W', '')

# Convertir las columnas a tipo float y manejar los valores vacíos (NaN)
bd_tw['Latitud'] = pd.to_numeric(bd_tw['Latitud'], errors='coerce')
bd_tw['Longitud'] = pd.to_numeric(bd_tw['Longitud'], errors='coerce')

# Convertir las coordenadas de longitud a valores negativos (oeste)
bd_tw['Longitud'] = -bd_tw['Longitud']

# Imprimir el DataFrame con los cambios
print(bd_tw)
```

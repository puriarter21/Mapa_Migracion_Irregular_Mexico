```python
import pandas as pd
import geopandas as gpd
from unidecode import unidecode
```


# Cargar y Preparar los Datos

Cargar el archivo CSV y el archivo GeoJSON usando geopandas y pandas. El archivo CSV debe de tener una columna que identifica cada estado (por ejemplo, "Estado") y el archivo GeoJSON una propiedad que se puede usar para hacer el match con los estados en el CSV.

```python
# Cargar el archivo CSV
datos_csv = pd.read_csv('/Users/pablouriarte/Documents/1. Expediente Tec de Monterrey/1.Tesis/Mapa_Migracion_Irregular_Mexico/4. mapas/datos_mapas/datos_inm_2022.csv')

# Cargar el archivo GeoJSON
datos_geojson = gpd.read_file('/Users/pablouriarte/Documents/1. Expediente Tec de Monterrey/1.Tesis/Mapa_Migracion_Irregular_Mexico/4. mapas/1. arcgis/maps_boundaries/mexico-with-regions_.geojson')
```

### Podria ser necesario normalizar los nombres para hacer el merge



```python
#En este caso utilizamos name y NOMGEO, pero nos aseguramos que coincidan quitando acentos

# Normalizar los nombres en el DataFrame y el GeoDataFrame
datos_csv['NOMGEO'] = datos_csv['NOMGEO'].apply(lambda x: unidecode(str(x)))
datos_geojson['name'] = datos_geojson['name'].apply(lambda x: unidecode(str(x)))
```

```python
#correcciones de nombre de los estados

# Función para normalizar los nombres y reemplazar valores específicos
def normalizar_y_reemplazar(nombre):
    # Normalizar quitando acentos
    nombre_normalizado = unidecode(nombre)
    # Reemplazar "Distrito Federal" por "Ciudad de México"
    if nombre_normalizado == "Distrito Federal":
        return "Ciudad de Mexico"
    else:
        return nombre_normalizado

# Aplicar la función a las columnas relevantes de cada DataFrame
datos_csv['NOMGEO'] = datos_csv['NOMGEO'].apply(normalizar_y_reemplazar)
datos_geojson['name'] = datos_geojson['name'].apply(normalizar_y_reemplazar)
```

```python
# Realizar el merge
datos_combinados = datos_geojson.merge(datos_csv, left_on='name', right_on='NOMGEO', how='left')

datos_combinados.head(32)
```

### Guardamos el archivo para despues importar en ArcGis

En este primer caso solo estamos usando datos de 2022, sin embargo para el futuro quisieramos hacer un mapa dinamico con los datos de detenciones en el tiempo.


```python
# Convertir explícitamente a GeoDataFrame si es necesario
if not isinstance(datos_combinados, gpd.GeoDataFrame):
    datos_combinados = gpd.GeoDataFrame(datos_combinados, geometry='geometry_x')

# Ahora, intenta guardar de nuevo como GeoJSON
datos_combinados.to_file('/Users/pablouriarte/Documents/1. Expediente Tec de Monterrey/1.Tesis/Mapa_Migracion_Irregular_Mexico/4. mapas/1. arcgis/data_map_layers/layer_eventos_inm_2022.geojson', driver='GeoJSON')
```
